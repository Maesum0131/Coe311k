```python
## monte carlo homework approximation for pi
import random
    
def montecarlo_pi(number_of_shots):
    number_of_hits = 0
   
    
    for i in range (number_of_shots):
        x = random.uniform(-1 ,1)
        y = random.uniform(-1, 1)
        
        if x*y + y*y <1:
            number_of_hits += 1
        return 4*number_of_hits/ number_of_shots
    
        
```

```python
import numpy
import math
import matplotlib.pyplot as plt
import random

shots = []
errors = []

true_pi = round(math.pi, 15)

for number_of_shots in range(10, 1001, 20):
    calculated_pi = montecarlo_pi(number_of_shots)
    true_error = abs(calculated_pi - true_pi)
    
    shots.append(number_of_shots)
    errors.append(true_error)
    
    print(number_of_shots, calculated_pi, true_error)
```

```python
plt.plot(shots, errors, marker='o')

plt.xlabel("Number of Shots")

plt.ylabel("Absolute Error")

plt.title("Monte Carlo Pi Approximation Error")

plt.grid(True)

plt.show()

```
