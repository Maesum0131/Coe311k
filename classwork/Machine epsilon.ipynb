```python
import numpy as np

pi_approx = 0.0
 
for n in range (5000):
    term = (-1)**n / (2*n + 1)
    pi_approx += term
if (n + 1)% 100 == 0 :
    pi_estimate = 4 * pi_approx
    relative_error = abs(pi_estimate - np.pi)/ np.pi

print(f"Terms: {n+1:4d}, π estimate: {pi_estimate:.8f}, Relative error: {relative_error:.8e}")
```
